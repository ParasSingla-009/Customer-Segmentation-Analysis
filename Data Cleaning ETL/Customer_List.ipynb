```python
drop_index_dob = newcustdf[newcustdf['DOB'].isnull()].index
drop_index_dob
```

```python
newcustdf.drop(drop_index_dob, inplace=True, axis = 0)
```

```python
newcustdf['DOB'].isnull().sum()
```

```python
newcustdf['DOB'] = pd.to_datetime(newcustdf["DOB"])
```

```python
def age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
newcustdf['Age'] = newcustdf["DOB"].apply(age)
```

```python
newcustdf["Age"].describe()
```

```python
plt.figure(figsize=(15,8))
sns.displot(newcustdf['Age'], bins = 50)
```

```python
newcustdf['Age Group'] = newcustdf['Age'].apply(lambda x : (math.floor(x/10)+1)*10)
```

```python
plt.figure(figsize=(15,8))
sns.displot(newcustdf['Age Group'], bins = 50)
```
